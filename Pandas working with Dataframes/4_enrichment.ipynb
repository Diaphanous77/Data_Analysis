# All

Прочитайте файл auto.json
- один из столбцов имеет тип float, поэтому давайте определим его формат в pandas с помощью pd.options.display float_format: числа с плавающей точкой должны отображаться с двумя десятичными знаками

```python
pd.options.display.float_format = '{:.2f}'.format
file_path = 'data/auto.json'
df = pd.read_json(file_path)
df
```

Обогатить фрейм данных, используя выборку из этого фрейма данных
- создать выборку из 200 новых наблюдений с random_state = 21
    * в выборке не должно быть новых комбинаций номера автомобиля, марки и модели, поэтому весь набор данных будет согласованным в этих терминах
    * нет ограничений по возврату и штрафам, вы можете взять любое значение из этих столбцов наугад и использовать его для любого номера автомобиля
- объединить выборку с исходным фреймом данных в новый фрейм данных concat_rows

```python
sample_size = 200

sample = df.sample(n=sample_size, replace=True, random_state=21)
sample['Refund'] = np.random.choice(df['Refund'].values, size=sample_size, replace=True)
sample['Fines'] = np.random.choice(df['Fines'].values, size=sample_size, replace=True)

concat_rows = pd.concat([df, sample], ignore_index=True)
concat_rows
```

Обогатить фрейм данных concat_rows новым столбцом с полученными данными
- создайте ряд с названием Год, используя случайные целые числа от 1980 до 2020
- используйте np.random.seed(21) перед генерацией лет
- соедините ряд с фреймом данных и назовите его Year

```python
np.random.seed(21)
years = np.random.randint(1980, 2021, size=concat_rows.shape[0])
year_series = pd.Series(years, name='Year')
concat_rows = pd.concat([concat_rows, year_series], axis=1)
concat_rows
```

Обогатить фрейм данных данными из другого фрейма данных
- получить самые популярные фамилии в США (файл surname.json)

```python
surnames_df = pd.read_json('data/surname.json')
surnames_df.columns = ['NAME', 'COUNT', 'RANK']
surnames_df
```

создайте новую серию с фамилиями (они не должны содержать специальных символов, таких как запятые, скобки и т. д.) из собранных вами данных, количество должно быть равно количеству уникальных номеров автомобилей, использующих выборку (используйте random_state = 21)

```python
np.random.seed(21)
car_numbers = concat_rows['CarNumber'].unique()
sample_surnames = np.random.choice(surnames_df['NAME'].values, size=len(car_numbers), replace=True)
sample_surnames
```

создайте датафрейм owners с 2 столбцами: CarNumber и SURNAME

```python
owners = pd.DataFrame({
    'CarNumber': car_numbers,
    'SURNAME': sample_surnames
})
owners
```

Добавьте еще 5 наблюдений к фрейму данных о штрафах

```python
makes_and_models = {
    'Honda': ['Civic', 'Accord', 'CR-V'],
    'BMW': ['X5', '320i', '530i'],
    'Audi': ['A4', 'A6', 'Q7']
}
car_numbers = ['X999XX77RUS', 'Y123AB45RUS', 'Z456CD78RUS', 'A789AB12RUS', 'B987XX45RUS']
refund = np.random.randint(1, 5, len(car_numbers))
fines = np.random.randint(100, 5001, len(car_numbers))
years = np.random.randint(1980, 2021, 5)
makes = np.array(list(makes_and_models.keys()))
make_indices = np.random.choice(len(makes), len(car_numbers))
selected_makes = makes[make_indices]
models = [
    np.random.choice(makes_and_models[make]) for make in selected_makes
]
new_fines_data = {
    'CarNumber': car_numbers,
    'Refund': refund,
    'Fines': fines,
    'Make': selected_makes.tolist(),
    'Model': models,
    'Year': years
}
new_fines_df = pd.DataFrame(new_fines_data)
fines = pd.concat([concat_rows, new_fines_df], ignore_index=True)
fines
```

удалите из датафрейма последние 20 наблюдений от владельцев и добавьте 3 новых наблюдения (они не совпадают с теми, которые вы добавляете в датафрейм штрафов)

```python
owners = owners[:-20]
new_owners_data = {
    'CarNumber': ['P123AB45RUS', 'Q456XX89RUS', 'R789AB12RUS'],
    'SURNAME': np.random.choice(surnames_df['NAME'].values, 3)
}
new_owners_df = pd.DataFrame(new_owners_data)

owners = pd.concat([owners, new_owners_df], ignore_index=True)
owners
```

Объедините оба фрейма данных
- новый фрейм данных должен содержать только те номера автомобилей, которые существуют в обоих фреймах данных

```python
common_df = pd.merge(fines, owners, on='CarNumber', how='inner')
common_df
```

новый фрейм данных должен содержать все номера автомобилей, которые существуют в обоих фреймах данных

```python
full_df = pd.merge(fines, owners, on='CarNumber', how='outer')
full_df
```

новый фрейм данных должен содержать только номера автомобилей из фрейма данных о штрафах

```python
fines_only_df = pd.merge(fines, owners, on='CarNumber', how='left')
fines_only_df
```

новый фрейм данных должен содержать только номера автомобилей из фрейма данных владельцев

```python
owners_only_df = pd.merge(fines, owners, on='CarNumber', how='right')
owners_only_df
```

Создайте сводную таблицу из фрейма данных штрафов

```python
pivot_table = pd.pivot_table(
    fines,
    values='Fines',
    index=['Make', 'Model'],
    columns='Year',
    aggfunc='sum',
    fill_value=np.nan,
)
pivot_table
```

Сохранить данные о штрафах и владельцах в CSV-файлы без индекса

```python
fines.to_csv('data/fines.csv', index=False)
owners.to_csv('data/owners.csv', index=False)
```
