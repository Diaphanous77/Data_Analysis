```python
import pandas as pd
import sqlite3
```

```python
connection = sqlite3.connect('data/checking-logs.sqlite')
```

анализировать только пользователей, а не администраторов
анализировать только даты, когда были и просмотры, и подтверждения проверки

```python
pageviews_query = '''
SELECT 
    date(datetime) AS date,
    COUNT(*) AS views
FROM pageviews
WHERE uid NOT LIKE 'admin_%'
GROUP BY date(datetime);
'''
pageviews_data = pd.read_sql(pageviews_query, connection)
pageviews_data
```

```python
checker_query = '''
SELECT 
    date(timestamp) AS date,
    COUNT(*) AS commits
FROM checker
WHERE uid NOT LIKE 'admin_%'
GROUP BY date(timestamp);
'''
checker_data = pd.read_sql(checker_query, connection)
checker_data
```

```python
merged_data = pd.merge(pageviews_data, checker_data, on='date', how='inner')
merged_data
```

```python
graph = merged_data.set_index('date').plot(
    figsize=(15, 8),
    fontsize=8,
    style={
        'views': 'r--o',
        'commits': 'c--'
    },
    title='Views and commits per day',
    rot=90
)
graph.set_xlabel('Date')
graph.grid(axis='x', linestyle='--')
```

```python
connection.close()
```

Сколько раз количество просмотров превышало 150?

```python
days_above_150 = (merged_data['views'] > 150).sum()
print(f'Ответ: {days_above_150}')
```
