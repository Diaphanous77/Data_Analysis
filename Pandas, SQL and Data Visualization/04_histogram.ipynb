```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
connection = sqlite3.connect('data/checking-logs.sqlite')
```

```python
checker_query = '''
SELECT 
    date(timestamp) AS date,
    time(timestamp) AS time,
    uid
FROM checker
WHERE uid NOT LIKE 'admin_%'
'''
checker_data = pd.read_sql(checker_query, connection)
checker_data
```

```python
def categorize_time_of_day(time_str):
    hour = int(time_str.split(':')[0])
    return hour
```

```python
checker_data['hour'] = checker_data['time'].apply(categorize_time_of_day)
checker_data['day_of_week'] = pd.to_datetime(checker_data['date']).dt.dayofweek
checker_data
```

```python
working_days_data = checker_data[checker_data['day_of_week'] < 5]
weekend_data = checker_data[checker_data['day_of_week'] >= 5]
```

```python
working_days_data
```

```python
weekend_data
```

```python
working_days_commits = working_days_data.groupby(['hour']).size()
weekend_commits = weekend_data.groupby(['hour']).size()
```

```python
working_days_avg = working_days_commits.groupby('hour').mean()
weekend_avg = weekend_commits.groupby('hour').mean()
```

```python
working_days_avg
```

```python
weekend_avg
```

```python
avg_commits = pd.DataFrame({
    'hour': working_days_avg.index,
    'weekend': weekend_avg.values,
    'working_day': working_days_avg.values,
})
avg_commits
```

```python
plt.figure(figsize=(15, 8))
plt.bar(
    avg_commits['hour'], 
    avg_commits['working_day'], 
    width=1.0, 
    label='working_days', 
    color='blue', 
    alpha=0.7, 
    align='edge'
)

plt.bar(
    avg_commits['hour'], 
    avg_commits['weekend'], 
    width=1.0, 
    label='weekends', 
    color='orange', 
    alpha=0.7, 
    align='edge'
)
plt.xlim(0, 24)

plt.xticks(range(0, 24, 5), fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)

plt.show()
```

Бывают ли часы, когда общее количество коммитов в выходные дни было выше, чем
в рабочие дни?

```python
higher_weekend_hours = avg_commits[avg_commits['weekend'] > avg_commits['working_day']].copy()

higher_weekend_hours['difference'] = higher_weekend_hours['weekend'] - higher_weekend_hours['working_day']

top_4_hours = higher_weekend_hours.sort_values('difference', ascending=False).head(4)

text = ''
for _, row in top_4_hours.iterrows():
    text += f'В {int(row['hour'])} часов: {int(row['weekend'])} коммитов в выходные и {int(row['working_day'])} коммитов в рабочие дни\n'
print(text)
```
